```python
import pandas as pd
```

```python
# Load data
df = pd.read_csv("consumer-price-indices_lka.csv"  )
```

```python
#missing value percentage
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
# Remove metadata row
df = df.iloc[1:].copy()
df.head()
```

```python
# Convert columns to appropriate types
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df['EndDate'] = pd.to_datetime(df['EndDate'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Months'] = df['Months'].astype(str)
```

```python
#drop columns
df.drop(columns=[ 'Iso3', 'Area Code', 'Area Code (M49)', 'Area', 'Element Code', 'Element', 'Months Code', 'Year Code', 'Unit', 'Flag', 'Note'], inplace=True)
```

```python
df.head()
```

```python
#rename columns
df.rename(columns={
    "StartDate": "Start_Date",
    "EndDate": "End_Date",
    "Item Code": "Item_Code",
    "Item": "Indicator",
    "Months": "Month",
    "Value": "CPI_Value"
}, inplace=True)
```

```python
df.head()
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
#duplicate
df.duplicated().sum()
```

```python
# Save to new CSV
df.to_csv("cleaned_cpi_sri_lanka.csv", index=False)
```
